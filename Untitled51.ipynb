```python
import pandas as pd
```

```python
historical['Date'] = pd.to_datetime(
    historical['Timestamp IST'], format="%d-%m-%Y %H:%M"
).dt.date
```

```python
sentiment['Date'] = pd.to_datetime(sentiment['date']).dt.date

```

```python
merged_df = pd.merge(
    historical,
    sentiment[['Date', 'classification']],
    on='Date',
    how='inner'
)
```

```python
avg_pnl = merged_df.groupby('classification')["Closed PnL"].mean().sort_values()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x=avg_pnl.index, y=avg_pnl.values, palette="coolwarm")
plt.title(" Average Trader PnL by Market Sentiment")
plt.xlabel("Sentiment Classification")
plt.ylabel("Average Closed PnL")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
trade_distribution = merged_df.groupby(['classification', 'Side']).size().unstack().fillna(0)
trade_distribution_percent = trade_distribution.div(trade_distribution.sum(axis=1), axis=0) * 100
```

```python
trade_distribution_percent.plot(
    kind='bar', stacked=True, figsize=(10, 6), colormap='Set2'
)
```

```python
plt.title("Trade Direction  Across Market Sentiments")
plt.ylabel("Percentage  Trades")
plt.xlabel("Sentiment")
plt.legend(title="Trade Side")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
avg_trade_size = merged_df.groupby('classification')['Size USD'].mean().sort_values()
```

```python
plt.figure(figsize=(10, 6))
sns.barplot(x=avg_trade_size.index, y=avg_trade_size.values, palette="viridis")
plt.title(" Avg Trade Size by Market Sentiment")
plt.xlabel("Sentiment")
plt.ylabel("Average Trade Size (USD)")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
account_pnl_sentiment = merged_df.groupby(['Account', 'classification'])['Closed PnL'].mean().unstack()
```

```python
top_accounts = account_pnl_sentiment.dropna().mean(axis=1).sort_values(ascending=False).head(10)
```

```python
top_accounts_df = account_pnl_sentiment.loc[top_accounts.index]

top_accounts_df.plot(
    kind='bar', figsize=(12, 6), colormap='coolwarm'
)
```

```python
plt.title(" Top 10 Trader Accounts: PnL Comparison by Sentiment")
plt.xlabel("Account")
plt.ylabel("Average Closed PnL")
plt.xticks(rotation=45)
plt.legend(title="Sentiment")
plt.tight_layout()
plt.show()
```
